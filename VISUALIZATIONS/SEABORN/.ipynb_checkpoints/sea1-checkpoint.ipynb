```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
# KDE for kernel density Estimation is a way to estimate the probability 
# density function of a continuous random variable 

# set the hist to false to yield a kde plot
sns.distplot(tips['total_bill'], hist=False)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
# MATRIX PLOTS 
tips.corr()
```

```python
# create a heatmap of correlation of the various variables
#  annotate them 
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=2)
```

```python
# CLUSTER MAP
#  hierachical map showing variables that are similar to each other
# similar data is close to each other
# notice that the order has changed for years or months
# normalize the dataset using a std scale
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```
