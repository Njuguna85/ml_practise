```python
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.set_style('darkgrid')
pg= sns.PairGrid(iris)
pg.map(plt.scatter)
```

```python
pg.map_diag(sns.distplot)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

```python
tips=  sns.load_dataset('tips')
gr = sns.FacetGrid(data=tips, col='time', row='smoker')
#  the plot below has
#     : row for smokers on the top and non smokers below
#     : columns for time, lunch on left and dinner on the right
#     : total_bill is the variable being mapped
gr.map(sns.distplot, 'total_bill')
```

```python
gr = sns.FacetGrid(data=tips, col='time', row='smoker')
gr.map(plt.scatter, 'total_bill', 'tip')
```
