```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#  create a numpy array
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('Title')
```

```python
# Create mutliplots on the same canvas
#  the subplot takes in the number of rows, columns and the plot being refered to
plt.subplot(1, 2, 1)
plt.plot(x,y, 'r')

plt.subplot(1,2, 2)
plt.plot(y,x, 'b')
```

```python
# OO METHOD
fig = plt.figure()

# the add axes takes in a list of left, bottom, width and height that range 0 to 1
axes = fig.add_axes([.1, .1, .8, .8])
# plot
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
```

```python
# Mutliple Plots
fig = plt.figure()
axes1 = fig.add_axes([.1, .1, .8, .8])
axes2 = fig.add_axes([.2, .16, .4, .5])

axes1.plot(x, y)
axes1.set_title('BIGGER')

axes2.plot(y, x)
axes2.set_title('INNER')
```

```python
# `             CREATING SUBPLOTS

# subplots will automatically call fig.add_axes
# axes is a list of matplotlib axes 
fig,axes = plt.subplots(nrows = 1, ncols = 2)
# axes.plot(x,y)

for curr_axe in axes:
    curr_axe.plot(x,y)

axes[1].plot(y,x)
axes[0].set_title('1st plot')

plt.tight_layout()
```

```python
#     FIGURE SIZE AND DPI
fig = plt.figure(figsize=(8,2), dpi=100)

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(8,2), dpi=100, nrows=2, ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#      ADD A LEGEND
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('TITLE')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.plot(x,x**2, label='X^2')
ax.plot(x,x**3, label='X^3')

ax.legend(loc=0)
```
